# One-way ANOVA to compare means of different departments

```python
from sklearn.feature_selection import f_oneway


f_statistic, p_value = f_oneway(emp_update[emp_update['Dept'] == 'HR']['rating'],
                                emp_update[emp_update['Dept'] == 'Technology']['rating'],
                                emp_update[emp_update['Dept'] == 'Sales']['rating'],
                                emp_update[emp_update['Dept'] == 'Marketing']['rating'])

print("\nHypothesis Testing Results:")
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

# Check the p-value to determine the significance of the results

```python
if p_value < 0.05:
    print("The null hypothesis is rejected. There is a significant difference in performance scores between different departments.")
else:
    print("The null hypothesis cannot be rejected. There is no significant difference in performance scores between different departments.")
```
